```python
import sys
sys.path.append("../")
from VQD_CAFQA import *
import numpy as np
import matplotlib.pyplot as plt
import scipy
from qiskit.quantum_info import Statevector
dlist = np.linspace(0.5,4,50)
%matplotlib inline

def getH4(d):
    return JWmolecule(f"H 0 0 0; H 0 0 {d}; H 0 0 {2*d}; H 0 0 {3*d}")
```

```python
print("----------------H4----------------")
H4Hamiltonian_Data=[]
for m,d in enumerate(dlist):
    coeffs,paulis,_ = getH4(d)
    D=np.linalg.eigh(SparsePauliOp(coeffs=coeffs,data=paulis))
    l=[[D[0][i],D[1].T[i]] for i in range(len(D[0]))]
    l.sort(key = lambda x:x[0])
    H4Hamiltonian_Data.append( (np.array([i[0] for i in l]),np.array([i[1] for i in l]).T))
    print(f"Distance {d}")
```

```python
from qiskit.quantum_info import Statevector

N_operator = N_op(8)
S2_operator=S2(8)
SZ_operator=S_z(8)

H4_singlet_energies = []
H4_triplet_energies = []
H4_cation_energies = []
H4_anion_energies = []

for m,i in enumerate(H4Hamiltonian_Data):
    singlet_l=[]
    triplet_l=[]
    cation_l=[]
    anion_l=[]
    
    s=i[1].T
    for k,j in enumerate(s):
        state=Statevector(j)
        N=round(state.expectation_value(SparsePauliOp(coeffs=N_operator.coeffs,data=[p[::-1] for p in N_operator.paulis.to_labels()])).real)
        SpinTotal=round(state.expectation_value(SparsePauliOp(coeffs=S2_operator.coeffs,data=[p for p in S2_operator.paulis.to_labels()])).real*4)/4
        SpinZ=round(state.expectation_value(SparsePauliOp(coeffs=SZ_operator.coeffs,data=[p[::-1] for p in SZ_operator.paulis.to_labels()])).real*2)/2
        if(N==4 and SpinTotal == 0 and len(singlet_l)<2):
            singlet_l.append(i[0][k].real)
        if(N==4 and SpinTotal == 2 and len(triplet_l)<2):
            if(len(triplet_l)==0 or i[0][k].real > max(triplet_l)+1e-5):
                triplet_l.append(i[0][k].real)
        if(N==5 and SpinTotal == 3/4 and len(cation_l)==0):
            cation_l.append(i[0][k].real)
        if(N==3 and SpinTotal == 3/4 and len(anion_l)==0):
            anion_l.append(i[0][k].real)
        if(len(singlet_l)==2 and len(triplet_l)==2 and len(cation_l)==1 and len(anion_l)==1):
            break
    H4_singlet_energies.append(singlet_l)
    H4_triplet_energies.append(triplet_l)
    H4_anion_energies.append(anion_l)
    H4_cation_energies.append(cation_l)
    print(f"Distance: {dlist[m]}")
```

```python
plt.title("H4 Molecule")
plt.plot(dlist,H4_singlet_energies,color='green',label='singlet')
plt.plot(dlist,H4_triplet_energies,color='orange',label='triplet')
plt.plot(dlist,H4_cation_energies,color='red',label='cation')
plt.plot(dlist,H4_anion_energies,color='blue',label='anion')
plt.legend()
```
